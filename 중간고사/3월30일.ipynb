```python
import numpy as np
```

```python
arr = np.array([[1.,2,3], [1,2,3]])
print(arr.dtype)
```

```python
arr1 = arr.astype(np.int8)
print(arr1, arr1.dtype)
```

```python
arr = np.array([[1.,2,3], [1,2,3]], dtype = 'int32')
print(arr.dtype)
```

```python
print(arr.shape)
```

```python
arr = np.array([[1.,2,3], [1,2,3]], dtype = 'int32')
print(arr.shape)
```

```python
arr1 = arr.reshape(1, 6)
print(arr1, arr1.shape)
print(arr, arr.shape)
```

```python
arr1 = arr.reshape(1, -1)
print(arr1, arr1.shape)
print(arr, arr.shape)
```

```python
arr1 = arr.reshape(-1, 1)
print(arr1, arr1.shape)
print(arr, arr.shape)
```

```python
arr = np.random.randn(3, 4)
print(arr, arr.shape)
```

```python
arr1 = arr.reshape(4, 3)
print(arr1, arr1.shape)
```

```python
arr2 = arr.reshape(-1, 2, 1)
print(arr2, arr2.shape)
```

```python
arr3 = arr.reshape(-1) # 1열로 쭉 피는거
print(arr3, arr3.shape)
```

```python
arr = np.random.randn(3, 4)
print(arr, arr.shape)
```

```python
arr3 = arr.reshape(1, 3, 4)
print(arr3, arr3.shape)
```

```python
arr4 = arr.reshape(1, 1, 3, 4) # 4차원
print(arr4, arr4.shape)
```

```python
arr5 = np.expand_dims(arr4, 3) # 5차원
print(arr5, arr5.shape)
```

## Matplotlib

```python
data = np.random.randn(50)
print(data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data)
plt.show
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(100)
#scat_data, hist_data
```

```python
plt.figure(figsize=(10000000000000000000., 5))
plt.subplot(2, 2, 1)
plt.plot(hist_data, 'r+')
plt.subplot(2, 2, 2)
plt.hist(hist_data, bins=20)
plt.subplot(2, 2, 3)
plt.scatter(scat_data, hist_data)
plt.show()
```

```python
hist_data = np.random.randn(10000)
plt.hist(hist_data, bins = 100)
plt.show()
```

```python
data = np.random.randn(100)
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(data, 'rx')
plt.show()
```

```python
plt.figure(figsize=(6.4, 4.8))
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(np.random.randn(100).cumsum(), 'r^', label = 'one')
plt.plot(np.random.randn(100).cumsum(), 'g.', label = 'two')
plt.plot(np.random.randn(100).cumsum(), 'b--', label = 'three')
plt.legend()
plt.savefig('graph.png')
plt.show()
```
