```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
```

```python
layer = tf.keras.layers.Conv2D(filters=8, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu')
```

```python
image8 = x_test[0:8]
```

```python
image8.shape, image8.dtype
```

```python
image8 = tf.cast(image8, dtype='float32')
```

```python
image8.dtype
```

```python
image8 = image8[..., tf.newaxis]
image8.shape
```

```python
img_out = layer(image8)
img_out.shape, np.min(img_out.numpy())
```
